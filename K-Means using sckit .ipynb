### Importing Liberaries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
plt.style.use('Solarize_Light2')
from warnings import filterwarnings
```

### loading dataset 

```python
X,y=make_blobs(n_samples=500,n_features=2,centers=5,random_state=3)
```

```python
X.shape,y.shape
```

```python
X[0],y[0]
```

```python
plt.figure(figsize=(5,4))
plt.scatter(X[:,0],X[:,1],alpha=0.7,c=y)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
```

# <font color ='green'> K-Means Clustering </font> 

### Determining the optimal value of K (Elbow method)

```python
filterwarnings('ignore')
wcss=[]
for i in range(1,11):
    kmean=KMeans(n_clusters=i)
    kmean.fit(X,y)
    wcss.append(kmean.inertia_)
print(wcss)    
```

### Visualizing elbow

```python
plt.figure(figsize=(5,3))
plt.grid()
plt.plot(range(1,11),wcss,linestyle='--',marker='o',color='b')
plt.xticks(range(1,11))
plt.xlabel('K values')
plt.ylabel('WCSS')
plt.title('Elbow method')
plt.show()
```

Here from the elbow method optimal value for the k should be 5 after 5 wcss values are not varying(not much reducing)

### Model buiding

```python
kmean=KMeans(n_clusters=5)
kmean.fit(X,y)
```

```python
centers=kmean.cluster_centers_
centers
```

```python
plt.figure(figsize=(6,4))
plt.scatter(X[:,0],X[:,1],alpha=0.7,c=y,s=70)
plt.plot(centers[:,0],centers[:,1],'^',color='r')
plt.grid(color='k', linestyle='--', linewidth=1)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title("K-Means Clustering")
```
